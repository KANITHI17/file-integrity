```python
import socket
import requests

# ------- Port Scanner -------
class PortScanner:
    def __init__(self, host, ports=range(1, 1025)):
        self.host = host
        self.ports = ports

    def scan(self):
        print(f"üîç Scanning ports on {self.host}")
        for port in self.ports:
            try:
                with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
                    s.settimeout(0.3)
                    result = s.connect_ex((self.host, port))
                    if result == 0:
                        print(f"[+] Port {port} is OPEN")
            except Exception as e:
                print(f"Error scanning port {port}: {e}")

# ------- Brute-force Directory Scanner -------
class BruteForceScanner:
    def __init__(self, host, wordlist=None):
        self.host = host
        self.wordlist = wordlist or ["admin", "login", "dashboard", "test", "uploads"]

    def run(self):
        print(f"üìÇ Running brute-force scan on {self.host}")
        for path in self.wordlist:
            url = f"http://{self.host}/{path}"
            try:
                res = requests.get(url, timeout=2)
                if res.status_code == 200:
                    print(f"[+] Found: {url}")
            except requests.RequestException:
                pass

# ------- Unified Runner -------
def run_tool(host, port_scan=False, brute_scan=False):
    if port_scan:
        scanner = PortScanner(host)
        scanner.scan()
    if brute_scan:
        brute = BruteForceScanner(host)
        brute.run()
```

```python
# ‚úÖ Test the toolkit
run_tool(
    host="example.com",        # Replace with your test target
    port_scan=True,            # Enable port scanner
    brute_scan=True            # Enable directory brute-forcing
)
```
